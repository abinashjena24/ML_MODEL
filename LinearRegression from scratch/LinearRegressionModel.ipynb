```python
import numpy as np
```

```python
x=np.array([1,2,3,4,5])
y=np.array([3,5,7,9,11])
#taking the equation y=2*x+1
```

```python
def prediction(x,theta0,theta1):
    return theta0+theta1*x
```

```python
def cost_est(x,y,theta0,theta1):
    m=len(x)
    predictions=prediction(x,theta0,theta1)
    cost=np.sum((predictions-y)**2)/(2*m)
    return cost
```

```python
def gradient_descent(x,y,theta0,theta1,learningrate,iteration):
    m=len(x)
    cost_his=[]
    for i in range(iteration):
        predictions=prediction(x,theta0,theta1)
        error=predictions-y
        del_theta0=np.sum(error)/m
        del_theta1=np.sum(error*x)/m
        theta0=theta0-(learningrate*del_theta0)
        theta1=theta1-(learningrate*del_theta1)
        cost=cost_est(x,y,theta0,theta1)
        cost_his.append(cost)
    return theta0,theta1,cost_his   
```

```python
theta0=0
theta1=0
theta0,theta1,cost_history=gradient_descent(x,y,theta0,theta1,learningrate=0.01,iteration=1000)
print("theta0=",theta0)
print("theta1=",theta1)
new_predictions=prediction(6,theta0,theta1)
print("predtiction for x=6",new_predictions)
print(cost_history[991:1001])
```

```python
y_pred = prediction(x, theta0, theta1)
mae = np.mean(np.abs(y - y_pred))

print("Mean Absolute Error (MAE):", mae)
```

```python
mse = np.mean((y - y_pred) ** 2)

print("Mean Squared Error (MSE):", mse)
```

```python
ss_res = np.sum((y - y_pred) ** 2)

ss_tot = np.sum((y - np.mean(y)) ** 2)

r2 = 1 - (ss_res / ss_tot)

print("R² Score:", r2*100,"%")
```

```python
import matplotlib.pyplot as plt

predictions = prediction(x, theta0, theta1)

plt.figure(figsize=(8,5))

plt.scatter(x, y, label="Actual Data")

plt.plot(x, predictions, label="Regression Line")

plt.xlabel("X")
plt.ylabel("Y")

plt.legend()
plt.grid(True)

plt.show()
```
